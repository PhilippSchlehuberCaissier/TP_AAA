# LTLf synthesis

LTLf is a "special" version of LTL working over **finite** traces.
We simulate this by translating them into büchi automata like so

```python
f = spot.formula("(i&X(i)) -> (X[4!](1) && G(i<->o))") # LTLf
ff = spot.from_ltlf(f) # LTL with special symbol
autf = spot.translate(ff, "buchi", "deterministic", "sbacc") # Translation
autf = spot.to_finite(autf) # DFA disguised as BDA
autf
```
